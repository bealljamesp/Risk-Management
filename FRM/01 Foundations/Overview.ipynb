Where You MUST Dig In (The Structural Foundations)There are three core concepts in Book 1 that are highly technical and form the absolute bedrock for your MSABA Capstone and future Risk Quant interviews. You must master these now:

The Classification of RisksYou must be able to instantly categorize risk types. This is not just trivia; it dictates the type of math you will use later.
Market Risk: Risk of losses due to movements in market prices (this is where our Beta, volatility, and linear projections live).
Credit Risk: Risk of default by a counterparty (requires probability of default models, transition matrices, and binary outcome math).
Operational Risk: Risk of loss from failed internal processes, people, systems, or external events (cyberattacks, system crashes). This uses fat-tailed loss distribution approaches (LDA) and Poisson processes.
Liquidity Risk: Split into Funding Liquidity (running out of cash) and Market Liquidity (not being able to sell an asset without tanking the price).

Modern Portfolio Theory (MPT) & The Capital Asset Pricing Model (CAPM)GARP introduces MPT and CAPM in Book 1 conceptually. This is where you should focus your energy. Because you already understand the vector mechanics of Beta, correlation ($\rho$), and the Market Risk Premium (MRP), this section will actually be enjoyable. Make sure you can write out and deconstruct the Security Market Line (SML) and understand the assumptions underlying efficient frontiers.

The GARP Code of ConductThis is a small but dangerous section. GARP always features specific, highly ambiguous ethical scenario questions on the exam. Read the Code of Conduct thoroughly once, focus on conflicts of interest and data integrity, and practice a few multiple-choice questions on Bionic Turtle to get used to their phrasing.

```python
📘 VaR Intuition Sheet (High‑Yield Relationships)
This is the sheet FRM actually tests — not formulas.

1. Volatility → VaR
VaR is directly proportional to volatility.

If volatility doubles, VaR roughly doubles.

GARCH spikes → VaR spikes.

Mental model:
Volatility is the “fuel” of VaR.

2. Correlation → Portfolio VaR
Higher correlation → higher portfolio VaR.

Lower correlation → diversification benefit → lower VaR.

Negative correlation → VaR can drop dramatically.

Key insight:
Correlation is the most powerful VaR lever.

3. Fat Tails → VaR Underestimation
Normal VaR underestimates risk in fat‑tail distributions.

t‑distribution → heavier tails → VaR too low.

ES captures this better.

FRM loves this.

4. Expected Shortfall vs VaR
ES > VaR in fat‑tail distributions.

ES is coherent; VaR is not.

ES accounts for tail severity; VaR only tail threshold.

Exam pattern:
“Which measure better captures tail risk?” → ES.

5. Time Horizon Scaling
Square‑root‑of‑time rule:

1‑day VaR → 10‑day VaR = ×√10

Only valid under:

i.i.d. returns

normality

no autocorrelation

FRM trick:
If volatility clusters → rule breaks.

6. Delta‑Normal VaR Limitations
Fails when:

Non‑linear products (options)

High gamma

Fat tails

Volatility clustering

Why:
It assumes linearity + normality.

7. Historical VaR Strengths/Weaknesses
Strengths:

No distribution assumptions

Captures fat tails

Easy to explain

Weaknesses:

Backward‑looking

Regime‑dependent

Sparse tail data

8. Monte Carlo VaR
Best when:

Non‑linear portfolios

Complex derivatives

Path‑dependent products

Weakness:

Model risk (garbage in → garbage out)

9. Liquidity → VaR
Wider bid‑ask spreads → higher VaR

Market depth drops → VaR spikes

Liquidity VaR = Market VaR + Liquidity add‑on

FRM loves liquidity add‑on questions.

10. Stress VaR
Uses stressed parameters (vol, correlation)

Basel requires it

Captures crisis‑mode behavior

Key idea:
Stress VaR ≠ historical VaR.
```

```python
Basel Framework — One‑Page Summary (FRM‑Optimized)
1. Purpose of Basel
The Basel framework exists to ensure banks hold enough capital, liquidity, and risk controls to survive stress events.
It governs:

Credit risk

Market risk

Operational risk

Liquidity risk

Basel evolves because models fail in the tails — FRM loves this theme.

2. Basel I → II → III → IV (The Evolution)
Basel I (1988) — Simple, credit‑risk‑only
Introduced Risk‑Weighted Assets (RWA)

Capital requirement: 8% of RWA

Four broad risk buckets (0%, 20%, 50%, 100%)

Weakness: too coarse, no market/operational risk

Basel II (2004) — Three‑Pillar System
Pillar 1 — Minimum Capital Requirements
Covers:

Credit risk (Standardized or IRB)

Market risk

Operational risk (BIA, STA, AMA)

Pillar 2 — Supervisory Review
Internal capital adequacy assessment (ICAAP)

Supervisors evaluate model quality

Pillar 3 — Market Discipline
Public disclosures to reduce opacity

Weakness:

Over‑reliance on internal models → underestimated tail risk (2008 crisis)

Basel III (2010–2017) — Post‑crisis repair
Focus: capital quality, liquidity, leverage, buffers

Capital Enhancements
CET1 is king (highest‑quality capital)

Higher minimums:

CET1: 4.5%

Tier 1: 6%

Total capital: 8%

Capital Buffers
Capital Conservation Buffer: +2.5% CET1

Countercyclical Buffer: 0–2.5%

SIFI/G‑SIB Buffers: extra CET1 for systemic banks

Leverage Ratio
Non‑risk‑based backstop

Tier 1 / Exposure ≥ 3%

Liquidity Standards
LCR (30‑day stress survival)

NSFR (1‑year stable funding)

Market Risk Reform (FRTB)
Replaces VaR with Expected Shortfall

Stricter model approval

Better capture of tail risk

Basel IV (2017–2023) — “Output Floor” + Model Limits
Not officially called “Basel IV,” but FRM uses the term.

Key Changes
Output Floor:

Internal model RWA ≥ 72.5% of standardized RWA

Prevents banks from gaming models

Revised Standardized Approaches:

Credit risk: more granular risk weights

Operational risk: SMA replaces AMA

FRTB fully implemented

More conservative capital calculations

3. Basel Capital Stack (What Counts as Capital)
CET1 (Highest Quality)
Common shares

Retained earnings

AOCI adjustments

Must absorb losses first

AT1 (Additional Tier 1)
Perpetual bonds

Contingent convertibles (CoCos)

Tier 2
Subordinated debt

Limited loss‑absorbing ability

Total Capital = CET1 + AT1 + Tier 2

4. Basel Risk Types & Measurement Approaches
Credit Risk
Standardized Approach: risk weights by asset class

IRB Approaches:

Foundation IRB (bank estimates PD)

Advanced IRB (bank estimates PD, LGD, EAD)

Market Risk
FRTB replaces VaR with Expected Shortfall

Two approaches:

IMA (Internal Models Approach)

SA (Standardized Approach)

Operational Risk
Basel II: BIA, STA, AMA

Basel IV: SMA only (simple, income‑based)

Liquidity Risk
LCR: survive 30‑day stress

NSFR: stable funding over 1 year

5. High‑Yield Basel Exam Patterns
Basel III = capital quality + liquidity + leverage

Basel IV = model limits + output floor

FRTB = ES replaces VaR

AMA is gone (replaced by SMA)

CET1 is always the most important capital measure

LCR = short‑term survival; NSFR = long‑term stability

Supervisors distrust internal models → output floor

6. One‑Sentence Basel Summary
Basel evolved from simple credit‑risk rules (Basel I) to a full capital‑liquidity‑model‑risk regime (Basel III/IV) designed to prevent banks from underestimating tail risk and to ensure they survive stress events.
```
